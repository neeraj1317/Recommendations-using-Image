# Converting all resized images to numpy array

# Finding euclidean distance between first image and all other images.

```python
distance=[]
for array in range(0,len(images_list)):
    #print(images_list[array])
    eucli=np.sqrt(np.sum((np.multiply(images_list[0],images_list[array]))))
    distance.append([eucli,images_list[array]])
    
                   
```

# Sorting the distance

```python
distance=sorted(distance, key=lambda x: x[0])
distance
```

# Similarly sqaured distance of first image and other images

```python
sqaured_distance=[]
for array in range(0,len(images_list)):
    #print(images_list[array])
    sqaureucli=np.sum((np.multiply(images_list[0],images_list[array])))
    sqaured_distance.append([sqaureucli,images_list[array]])
```

# Sorting sqaured distance

```python
sqaured_distance=sorted(sqaured_distance, key=lambda x: x[0])
sqaured_distance
```

# FINDING AND STORING 10 NEAREST NEIGHBOURS TO IMAGE

```python
euclidean_knn = []
for i in range(0,10):
    img=Image.fromarray(distance[i][1],'RGB')
    euclidean_knn.append(img)
    for im in euclidean_knn:
        im.save(path+'/'+'closest.png')
```

```python
squared_euclidean_knn = []
for i in range(0,10):
    img=Image.fromarray(sqaured_distance[i][1],'RGB')
    squared_euclidean_knn.append(img)
    for im in squared_euclidean_knn:
        im.save(path+'/'+'sqauredclosest.png')
```
